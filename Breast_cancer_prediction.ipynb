```python
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
Labelencoder_Y=LabelEncoder()
df.iloc[:,1]=Labelencoder_Y.fit_transform(df.iloc[:,1].values)
```

```python
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
logr=LogisticRegression(random_state=0)
logr.fit(xtrain,ytrain)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(criterion='entropy',random_state=0)
tree.fit(xtrain,ytrain)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
forest.fit(xtrain,ytrain)
```

```python
print('acc',logr.score(xtrain,ytrain))
print('acc',tree.score(xtrain,ytrain))
print('acc',forest.score(xtrain,ytrain))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
cfm=confusion_matrix(ytest,logr.predict(xtest))
classification_report(ytest,tree.predict(xtest))


tp=cfm[0][0]
tn=cfm[1][1]
fp=cfm[0][1]
fn=cfm[1][0]
print('acc',(tp + tn)/(tp+tn+fp+fn))
```

```python
print(classification_report(ytest,logr.predict(xtest)))
print(accuracy_score(ytest,logr.predict(xtest)))
print(classification_report(ytest,tree.predict(xtest)))
print(accuracy_score(ytest,tree.predict(xtest)))
print(classification_report(ytest,forest.predict(xtest)))
print(accuracy_score(ytest,forest.predict(xtest)))
```

```python
from sklearn.svm import SVC
sv = SVC(kernel = 'linear', random_state = 0)
sv.fit(xtrain,ytrain)
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(xtrain, ytrain)
print(classification_report(ytest,sv.predict(xtest)))
print(accuracy_score(ytest,sv.predict(xtest)))
print(classification_report(ytest,gnb.predict(xtest)))
print(accuracy_score(ytest,gnb.predict(xtest)))

```
